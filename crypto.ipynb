```python
file_path = "Data/shopping_data.csv"
shopping_df = pd.read_csv(file_path)
shopping_df.head()
```

```python
# Columns
shopping_df.columns
```

```python
shopping_df.dtypes
```

```python
# Find null values
for column in shopping_df.columns:
    print(f"Column {column} has {shopping_df[column].isnull().sum()} null values")
```

```python
# Drop null rows
shopping_df = shopping_df.dropna()
shopping_df.head()
```

```python
# Find duplicate entries
print(f"Duplicate entries: {shopping_df.duplicated().sum()}")
```

```python
shopping_df.drop(columns=["CustomerID"], inplace=True)
shopping_df.head()
```

```python
# Transform String column
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0
    
shopping_df["Card Member"] = shopping_df["Card Member"].apply(change_string)
shopping_df.head()
```

```python
# Transform Annual Income
shopping_df["Annual Income"] = shopping_df["Annual Income"]/1000
shopping_df.head()
```

```python
shopping_df.rename(columns = {'Card Member' : 'CardMember', 'Annual Income' : 'AnnualIncome', "Spending Score (1-100)" : "SpendingScore"}, inplace = True)
shopping_df.head()
```

```python
# Saving cleaned data
file_path = "Data/shopping_data_cleaned.csv"
shopping_df.to_csv(file_path, index=False)
```
